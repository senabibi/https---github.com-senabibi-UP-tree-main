```python
def check_df(data_path):
    # Veriyi yükleme
    dataframe = pd.read_csv(data_path)

    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### Head #####################")
    print(dataframe.head(3))
    print("##################### Tail #####################")
    print(dataframe.tail(3))
    print("##################### NA #####################")
    print(dataframe.isnull().sum())
    print("##################### Quantiles #####################")
    print(dataframe['Ratio'].quantile([0, 0.05, 0.50, 0.95, 0.99, 1]))

# Veri çerçevesini kontrol etme
data_path = "/female_ratio.csv"
check_df(data_path)
```

Shape (Şekil): Veri çerçevesinin boyutunu (satır, sütun) belirtir. Bu veri çerçevesinde 69 satır ve 2 sütun bulunmaktadır.

Types (Türler): Her sütunun veri türünü belirtir. "Entity" sütunu metin (object) türünde iken, "Ratio" sütunu ondalık sayı (float64) türündedir.

Head (Baş): Veri çerçevesinin ilk üç satırını gösterir. Bu başlangıç ​​satırları, veri çerçevesinin genel yapısını anlamak için kullanılır.

Tail (Kuyruk): Veri çerçevesinin son üç satırını gösterir. Bu son satırlar, veri çerçevesinin sonundaki verileri incelemek için kullanılır.

NA (Eksik Değerler): Her sütundaki eksik değerlerin sayısını belirtir. Bu veri çerçevesinde herhangi bir eksik değer bulunmamaktadır.

Quantiles (Çeyreklikler): "Ratio" sütununun çeyrekliklerini ve diğer yüzdelerini gösterir. Bu, veri dağılımının genel görünümünü anlamak için kullanılır. Örneğin, verinin %95'lik dilimindeki en yüksek oran 7.378'dir.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Veriyi yükleme
data = pd.read_csv("/female_ratio.csv")

# Oranları bir listeye ekleyerek histogram oluşturma
ratios = data['Ratio'].tolist()

# Histogram oluşturma
plt.figure(figsize=(10, 6))
plt.hist(ratios, bins=20, color='red', edgecolor='black')  # 20 bölüme sahip bir histogram oluşturuldu
plt.title('2014 Yılında Ülkelerde Kadın/Erkek Ücretsiz Bakım İşgücü Oranı Dağılımı')
plt.xlabel('Kadın/Men Oranı')
plt.ylabel('Frekans')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
